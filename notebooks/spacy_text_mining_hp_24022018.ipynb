### Phase 2 : Import dataset

```python
#Specify path where the data is.
DATA_PATH = '../data/raw/'

#Specify the name fo the source file.
DATA_FILE = 'hp1.txt'
```

```python
#Open connection to file, read the text and close it.
file = open(DATA_PATH + DATA_FILE, 'rt', encoding='utf8')
text = file.read()
file.close()
```

### Phase 3 : Generate spaCy tags

```python
#Load spacy english model
nlp = spacy.load('en')

#Tag
docs = nlp(text)
```

```python
#NER and detect PERSON tags
person_list = []
for token in docs.ents:
    if(token.label_ == 'PERSON'):
        person_list.append(token)
```

```python
#Use inbuilt renderer to check tags
displacy.render(docs, style='ent', jupyter=True)
```
